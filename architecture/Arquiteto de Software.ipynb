## Definição e Elementos da Arquitetura
> Arquitetura diz respeito a coisas importantes, seja lá o que importante signifique - Martin Fowler

Arquitetura é a coordenação de 4 dimensões:
* __Caraterísticas Arquiteturais:__ Definem critérios de sucesso (disponibilidade, confiabilidade, segurança...)
* __Estrutura:__ Se refere ao padrão arquitetural implementado (microservices, camadas, microkernel...)
* __Decisões Arquiteturais:__ Definem as regras de como o sistema deve ser construído
* __Princípios do Desenho:__ Atuam como diretriz à equipe de desenvolvimento

## Características Arquiteturais
Também conhecidas como 'Requisitos não funcionais" ou "Requisitos Qualitativos"

Características Arquiteturais atendem 3 critérios:
1. Especificam considerações do desenho não ligadas ao domínio de negócios.
2. Influenciam algum aspecto estrutural do desenho.
3. São críticos para o sucesso da aplicação.

Exemplos:
* __Eficiência:__ Utilização máxima de CPU deverá ser 90%"
* __Confiabilidade:__ O tempo médio para recuperação deverá ser 3 minutos
* __Segurança:__ Os dados necessitarão de autorização prévia

Padrões Arquiteturais descrevem um relacionamento entre componentes que cobram uma variedade de características arquiteturais.

Exemplos de Padrões Arquiteturais:
* __Big ball of mud:__ total ausência de estrutura
* __Client/Server:__ Primeira separação da arquitetura original de computadores. Franciona a aplicação em 1 provedor de informações (Server) e 1 ou mais consumidores (Clients)

Exemplos de estilos arquiteturais:
1. __Monolítico:__
    * Arquitetura em Camadas
    * Arquitetura em Pipeline
    * Arquitetura de MicroKernel
2. __Distribuído:__
    * Arquitetura baseada em serviços
    * Arquitetura baseada em microserviços
    * Arquitetura orientada a eventos

# Capítulo 2 - O Papel do Arquiteto
> O arquiteto deve ser como um guia, um experiente e capacitado membro da equipe que ensina aos outros se virar melhor, e ainda assim está sempre lá para as partes mais complicadas.

## Expectativas em relação ao papel de Arquiteto

* __Tomar decisões arquiteturais__
    - O arquiteto deve ser aquele que foca em guiar a tomada de decisão.
* __Analisar a arquitetura continuamente__
    - Manter a relevância da solução.
* __Manter-se atualizado com as tendências__
    - A capacidade de se manter atualizado é o que garante arquiteturas duradouras.
* __Garantir conformidade com decisões__
    - Garantir que as decisões tomadas sejam cumpridas.
* __Diversificar entre experimentação e exposição__
    - Buscar o equilíbrio entre __as soluções que o arquiteto conhece__ e __as soluções que o arquiteto domina__
* __Ter conhecimento de negócio__
    - Buscar sempre entender os problemas e objetivos do negócio, visando uma solução efetiva.
* __Possuir habilidades interpessoais__
    - Somente consegue viabilizar seu trabalho através de características como comunicação, facilitação, liderança, trabalho em equipe e negociação.
* __Entender e navegar na política__
    - Saber utilizar técnicas de negociação para conseguir aprovações.

## Mentalidade de Arquitetura

O que molda a mentalidade de um arquiteto:
* __Estabelecer interfaces entre arquitetura e desenho__
* __Compatibilizar visão ampla X visão profunda__
    - Conseguir discernir entre o que se sabe e o que não se sabe
* __Analisar trade-offs__
* __Entender fatores de negócio__


# Capítulo 3 - Abordagens do Desenvolvimento #
> São as formas de se encarar a missão de construir e manter as soluções desde a criação até sua descontinuidade.

## Modelos de Ciclo de Vida ##
Definição das fases que conectam o início e o fim de um produto.

Exemplos de modelos:
* __Cascata:__ 
    * Também conhecido como Clássico ou Linear.
    * sugere uma abordagem sequencial começando pela nível de sistema e progredindo pelas fases de análise >> design >> codificação >> testes e suporte
    * Problemas conhecidos:
        * Projetos reais raramente são lineares;
        * É usualmente difícil estabelecer todos os requisitos explicitamente, principalmente nas fases iniciais;
        * Entrega da versão funcional somente no fim do projeto.
* __Espiral:__
    * É um modelo evolucionário, utiliza-se de prototipação aliada aos aspectos controlados do modelo __cascata__
    * Provê um potencial de rápido desenvolvimento de versões incrementais
* __Incremental:__
    * Obtida de forma estagiada à medido em que o tempo passa
    * Cada sequência linear produz uma entrega incremental utilizável de software

## Abordagem Ágil ##
__Manifesto:__
* _Indivíduos e Interações_ mais que processos e ferramentas
* _Software em funcionamento_ mais que documentação abrangente
* _Colaboração com o cliente_ mais que negociação de contratos
* _Responder a mudanças_ mais que seguir um plano

__12 Princípios do Manifesto Ágil__
1. Nossa maior prioridade é satisfazer o cliente, através da entrega adiantada e contínua de software de valor
2. Aceitar mudanças de requisitos, mesmo no fim do desenvolvimento. Processos ágeis se adequam a mudanças para que o cliente possa tirar vantagem competitiva
3. Entregar software funcionando com frequência, na escala de semanas até meses, com preferência aos períodos mais curtos
4. Pessoas relacionadas à negócios e desenvolvedores devem trabalhar em conjunto diariamente, durante todo o custo do projeto
5. Construir projetos ao redor de indivíduos motivados, dando a eles o ambiente e suporte necessário confiando que farão seu trabalho
6. O método mais eficiente e eficaz de transmitir informações para e por dentro de um time é através de uma conversa cara a cara
7. Software funcional é a medida primária de progresso
8. Processos ágeis promovem um ambiente sustentável. Os patrocinadores, devs e usuários devem ser capazes de manter indefinidamente passos constantes
9. Contínua atenção às excelência técnica e bom design aumentam a agilidade
10. Simplicidade: a arte de maximizar a quantidade de trabalho __que não precisou ser feito__
11. As melhores arquiteturas, requisitos e designs emergem de times auto-organizáveis
12. Em intervalos regulares, o time reflete em como ficar mais efetivo, então, se ajustam e otimizam seu comportamento de acordo

### DevOps ###
> Movimento emergente que preconiza um trabalho colaborativo entre DevTeam e OpsTeam.
* __Princípios:__
    * Fluxo: 
        * Objetivo: Reduzir o tempo necessário para as mudanças serem disponibilizadas em produção, garantindo qualidade e confiabilidade
        * Práticas:
            * Criação fundações do pipeline de deployment
            * Habilitar e praticar Continuous Integration
            * Habilitar testes automatizados rápidos e confiáveis
            * Definir arquitetura, automatizar e habilitar releases de baixo risco
    * Feedback:
        * Objetivo: Criar um sistema de trabalho cada vez mais resiliente e seguro
        * Práticas:
            * Identificar os problemas quando ocorrem
            * Criar telemetria para identificação e solução de problemas
    * Cultura de Aprendizagem e Experimentação Contínua
        * Objetivo: Criar cultura de alta confiança, reforçando a condição de _long life learners_, que devem assumir riscos
        * Práticas:
            * Institucionalizar melhoria contínua do trabalho
            * Habilitar e injetar aprendizagem no trabalho diário
            
* __Pipeline de Deployment:__
    * Objetivos: 
        * Criar todo o ambiente de produção completamente automatizado a partir do controle de versões
        * Prover feedback mais rápido possível se a mudança está passível de entrega de valor (funcionando como esperado)

* __Sistema de Controle de Versões:__
    * Objetivos:
        * Garantir acesso controlado a cada versionamento do código trabalho
        * Permitir o trabalho colaborativo entre times ao longo do tempo
        * Rastreabilidade de colaboradores

## Capítulo 1 - O que é e para que serve a Engenharia de Requisitos
* Considerada a parte mais crítica do projeto, impacto direto na qualidade da solução.
* Requisitos guiam as atividades do projeto, é a base para todo o projeto.

**Requisitos definem a base para:**
    * Planejamento do Projeto (Coleta)
    * Gerenciamento dos riscos (Análise)
    * Testes de Aceitação (Documentação)
    * Controle de Mudanças (Gerência)

## Capítulo 2 - Processo de Engenharia de Requisitos
**Ao dar início a um projeto, precisamos:**
    1. Levantar requisitos
    2. Entendê-los
    3. Documentá-los
**Etapas do Processo de Engenharia de Requisitos:**
    * Concepção - Identifica-se stakeholders e sua visão do problema
    * Elicitação - Levanta-se requisitos de usuário (Funcionais e Não funcionais
    * Elaboração - Detalha-se cada requisito em modelos conceituais (UML)
    * Negociação - Identificar conflitos para negociação com stakeholders (pririozar, eliminar, combinar)
    * Especificação - Detalha-se em termos técnicos o que se deve desenvolver
    * Validação - Valida-se a cobertura da solução em relação aos requisitos
    * Gerenciamento - Permeia todo o ciclo com objetivo de garantir o escopo e gerir mudanças
