# Enade DataSet

## Extraction

1. Steps:
    * Load libs
    * Make a directory
    * Download url (via requests)
    * Unzip
    * Check process

```python
# Loading Libs
import pandas as pd
import numpy as np
import zipfile
import requests as r
from io import BytesIO
import os

url: str = "http://download.inep.gov.br/microdados/Enade_Microdados/microdados_enade_2019.zip"
path: str = "../data/enade"
```

```python
# Make a directory
os.makedirs(path, exist_ok=True)

# Download
file_bytes: BytesIO = BytesIO(r.get(url).content)
    
# Unzip
zipfile.ZipFile(file_bytes).extractall(path)
```

```python
# Check downloaded file
file_path: str = path + '/microdados_enade_2019/2019/3.DADOS/microdados_enade_2019.txt'
df = pd.read_csv(file_path, sep=';', decimal=',')
df.head()
```
